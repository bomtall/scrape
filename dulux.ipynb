```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = "https://www.duluxdecoratorcentre.co.uk/dulux-trade-vinyl-matt"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
scripts = soup.find_all(name='script', attrs={"type": "text/javascript"})
```

```python
script = scripts[0]
print(script.prettify())
```

```python
scripts
```

```python
paint_data = json.loads(scripts[0].contents[0].split('function')[0].split('=')[-1].rstrip(',];\r\n ') + ']')
```

```python
paint_data[19]
```
